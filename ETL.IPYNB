```python
#Importing Packages
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

#setting location for CSV files to be replaced with Amazon AWS connection 
fp_1 = 'calendar.csv'
fp_2 = 'listings.csv'
fp_3 = 'reviews.csv'

#reading CSV files into DF
cal_df = pd.read_csv(fp_1)
list_df = pd.read_csv(fp_2)
rev_df = pd.read_csv(fp_3)

```

```python
#creates ML Dataframe inclusive of review info
ml_df = list_df[["reviews_per_month","require_guest_phone_verification","require_guest_profile_picture","cancellation_policy","instant_bookable","requires_license","review_scores_value","review_scores_location","review_scores_communication","review_scores_checkin","review_scores_cleanliness","review_scores_accuracy","review_scores_rating","number_of_reviews","availability_365","availability_90","availability_60","availability_30","maximum_nights","minimum_nights","extra_people","guests_included","price","bed_type","beds","bedrooms","bathrooms","accommodates","room_type","property_type","neighbourhood_group_cleansed","host_identity_verified","host_has_profile_pic","host_is_superhost","host_acceptance_rate","host_response_rate","host_response_time"]].copy()

# counting df
ml_df.head()
```

```python
#droping NAs in machine learning dataframe
ml_df = ml_df.dropna()
ml_df.count()
```

```python
#defining a good or a bad rating 
ml_df['rating_good_or_bad'] = np.where(ml_df['review_scores_value'] > 8, 'good_review', 'bad_review')

#checking good vs. bad rating count
good_bad_review_count = ml_df.groupby('rating_good_or_bad')['reviews_per_month'].count()
good_bad_review_count

```

```python
#removing rating info to avoid skewing ML model 
ml_final = ml_df[["rating_good_or_bad","require_guest_phone_verification","require_guest_profile_picture","cancellation_policy","instant_bookable","requires_license","availability_365","availability_90","availability_60","availability_30","maximum_nights","minimum_nights","extra_people","guests_included","price","bed_type","beds","bedrooms","bathrooms","accommodates","room_type","property_type","neighbourhood_group_cleansed","host_identity_verified","host_has_profile_pic","host_is_superhost","host_acceptance_rate","host_response_rate","host_response_time"]].copy()

```

```python
#encoding for all true and false variables: 

##encoding for superhost

# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')
# passing bridge-types-cat column (label encoded values of bridge_types)
enc_df = pd.DataFrame(enc.fit_transform(ml_final[['host_is_superhost']]).toarray())
# merge with main df bridge_df on key values
ml_final= ml_final.join(enc_df)
ml_final.rename(columns = {0:'not_a_superhost', 1:'a_superhost'}, inplace = True)
ml_final.head()


```

```python
##encoding for host has profile pic

# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')
# passing bridge-types-cat column (label encoded values of bridge_types)
enc_df = pd.DataFrame(enc.fit_transform(ml_final[['host_has_profile_pic']]).toarray())
# merge with main df bridge_df on key values
ml_final= ml_final.join(enc_df)
ml_final.rename(columns = {0:'host_does_not_have_profile_pic', 1:'host_does_have_profile_pic'}, inplace = True)
ml_final.head()
```
