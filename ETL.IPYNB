```python
#Importing Packages
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
import psycopg2
import sys
import boto3
import os
from config import config_database, config_password

##passowrd from config file 
pww = config_password

## AWS Database connection
conn = psycopg2.connect(database= "final-project",
                        host="final-project.cuwr3km7ozpl.us-east-1.rds.amazonaws.com",
                        user= "postgres",
                        password=pww ,
                        port="5432")

```

```python
## Creates Unit Dataframe: 

cur = conn.cursor()
cur.execute("SELECT * FROM unit")

tuples_list = cur.fetchall()
cur.close()
column_names_unit = ["id","host_id","booking_id","name","neighbourhood_group","city","state","zipcode","latitude","longitude","property_type","room_type","accommodates",
"bathrooms","bedrooms","beds","bed_type","instant_bookable","cancellation_policy","require_guest_profile_picture","require_guest_phone_verification","number_of_reviews","first_review",
"last_review","review_scores_rating","review_scores_accuracy","review_scores_cleanliness",
"review_scores_checkin","review_scores_communication","review_scores_location","review_scores_value","reviews_per_month"]

unit = pd.DataFrame(tuples_list, columns=column_names_unit)
unit.head()
```

```python
## Creates Host Dataframe: 

cur = conn.cursor()

cur.execute("SELECT * FROM host")

tuples_list = cur.fetchall()
cur.close()
column_names_host = ["host_id","host_name","host_since","host_listings_count","host_location","host_response_time","host_response_rate",
"host_acceptance_rate","host_is_superhost","host_has_profile_pic","host_identity_verified"]

host = pd.DataFrame(tuples_list, columns=column_names_host)
host.head()
```

```python
#Checks Unit Table: 

unit_rows = len(unit.axes [0])

unit_cols = len(unit.axes[1])

print(unit)
print("Number of Rows: ", unit_rows)
print("Number of Columns: ", unit_cols)
```

```python
#Checks host table

host_rows = len(host.axes [0])

host_cols = len(host.axes[1])

print(host)
print("Number of Rows: ", host_rows)
print("Number of Columns: ", host_cols)
```

```python
#merges both tables and checks output
#creates "ml_df"

ml_df = unit.merge(host, how = 'outer', on='host_id')

ml_df_rows = len(ml_df.axes [0])
ml_df_cols = len(ml_df.axes[1])

print(ml_df)
print("Number of Rows: ", ml_df_rows)
print("Number of Columns: ", ml_df_cols)
```

```python
#droping NAs in machine learning dataframe
ml_df = ml_df.dropna()
ml_df.count()
```

```python
#defining a good or a bad rating avg rating 9.46
ml_df['rating_good_or_bad'] = np.where(ml_df['review_scores_value'] > 8.9, 'good_review', 'bad_review')

#checking good vs. bad rating count
good_bad_review_count = ml_df.groupby('rating_good_or_bad')['reviews_per_month'].count()
print(good_bad_review_count)

#ml_df.to_csv('ml_check.csv', index= False)
```

```python
#removing rating info to avoid skewing ML model 
ml_final = ml_df.copy()
ml_final.reset_index(inplace=True)
ml_final
```

Encodes for All True and False Variables: 

```python
# Creates T/F encoding based on avg. number of reviews, accommodations, beds, and baths 

# number of review avg = 30
# number of accomodations avg = 3
# number of bedrooms avg = 1
# number of beds avg = 2
# number of baths avg = 1.5
# host response rate avg = 0.95
# host acceptance rate avg = .99

ml_final['num_of_reivews_gt_avg'] = np.where(ml_df['number_of_reviews'] > 30, True, False)
ml_final['num_of_accom_gt_avg'] = np.where(ml_df['accommodates'] > 3, True, False)
ml_final['num_of_bedrooms_gt_avg'] = np.where(ml_df['bedrooms'] > 1, True, False)
ml_final['num_of_beds_gt_avg'] = np.where(ml_df['beds'] > 2, True, False)
ml_final['num_of_baths_gt_avg'] = np.where(ml_df['bathrooms'] > 1.5, True, False)
ml_final['response_rate_gt_avg'] = np.where(ml_df['bathrooms'] > .95, True, False)
ml_final['accept_rate_gt_avg'] = np.where(ml_df['bathrooms'] > .99, True, False)

ml_final.head()


```

```python
##encoding for acceptance rate

# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')
# encoding values
enc_df = pd.DataFrame(enc.fit_transform(ml_final[['accept_rate_gt_avg']]).toarray())
# joining with database
ml_final= ml_final.join(enc_df)
ml_final.rename(columns = {0:'accept_rate_below_avg', 1:'accept_rate_above_avg'}, inplace = True)
ml_final.tail()
```
