```python
#Converting these Price object dtypes to Integers
listings_df["price", "weekly_price", "monthly_price", "security deposits", "extra people", "cleaning fee"] = (listings_df["price", "weekly_price", "monthly_price", "security deposits", "cleaning fee"]).replace("[$,]", "", regex=True).astype(int)
listings_df.dtypes()
```
